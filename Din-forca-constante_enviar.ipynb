```python
import numpy as np
import matplotlib.pyplot as plt

c = 3e8         # velocidade da luz (m/s)

# PRTON
# Constantes fsicas
q = 1.602e-19  # carga do prton (C)
m = 1.67e-27   # massa do prton (kg)
E = 1.0e0      # campo eltrico (V/m) (para o prton)

# Parmetros da simulao
dt = 1e-5      # passo de tempo (s) (para o prton)
t_max = 1e1    # tempo total (s) (para prton)
n_steps = int(t_max / dt)

"""
# ELTRON
# Constantes fsicas
q = -1.602e-19  # carga do eltron (C)
m = 9.109e-31   # massa do eltron (kg)
E = -1.0e0    # campo eltrico (V/m) (para o eltron)

# Parmetros da simulao
dt = 1e-9      # passo de tempo (s)
t_max = 1e-3    # tempo total (s) (para eltron)
n_steps = int(t_max / dt)
"""

# Inicializaes
t = np.linspace(0, t_max, n_steps)
v_classical = np.zeros(n_steps)
v_relativistic = np.zeros(n_steps)

# Integrao com mtodo de Euler
for i in range(1, n_steps):
    # Acelerao clssica (constante)
    a_classical = q * E / m
    v_classical[i] = v_classical[i-1] + a_classical * dt

    # Acelerao relativstica
    gamma = 1 / np.sqrt(1 - (v_relativistic[i-1]/c)**2) if v_relativistic[i-1] < c else 1e10  # evitar diviso por zero
    a_rel = (q * E) / (gamma**3 * m)
    v_relativistic[i] = v_relativistic[i-1] + a_rel * dt

# Plotagem
plt.figure(figsize=(10, 6))
plt.plot(t * 1e3, v_classical / 1e8, label='Clssica', linestyle='--')
plt.plot(t * 1e3, v_relativistic / 1e8, label='Relativstica')
plt.xlabel('Tempo (ms)')
plt.ylabel(r'Velocidade ($10^8$ m/s)')
plt.title('Velocidade do eltron sob um campo eltrico constante')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Prton
Foram ajustados os parmetros, para o clculo da velocidade do prton: 
- massa, em kg
- carga, em C

Foram tambm ajustados os seguintes parmetros:
- campo eltrico: para que o mesmo ficasse positivo, considerando a carga do prton.
- tempo, pois, por ser a massa do prton maior, ser mais demorado para atingir velocidades relativsticas.
- passo de tempo, devido  ordem de grandeza.
