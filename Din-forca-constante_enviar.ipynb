```python
import numpy as np
import matplotlib.pyplot as plt

c = 3e8         # velocidade da luz (m/s)

# PRÓTON
# Constantes físicas
q = 1.602e-19  # carga do próton (C)
m = 1.67e-27   # massa do próton (kg)
E = 1.0e0      # campo elétrico (V/m) (para o próton)

# Parâmetros da simulação
dt = 1e-5      # passo de tempo (s) (para o próton)
t_max = 1e1    # tempo total (s) (para próton)
n_steps = int(t_max / dt)

"""
# ELÉTRON
# Constantes físicas
q = -1.602e-19  # carga do elétron (C)
m = 9.109e-31   # massa do elétron (kg)
E = -1.0e0    # campo elétrico (V/m) (para o elétron)

# Parâmetros da simulação
dt = 1e-9      # passo de tempo (s)
t_max = 1e-3    # tempo total (s) (para elétron)
n_steps = int(t_max / dt)
"""

# Inicializações
t = np.linspace(0, t_max, n_steps)
v_classical = np.zeros(n_steps)
v_relativistic = np.zeros(n_steps)

# Integração com método de Euler
for i in range(1, n_steps):
    # Aceleração clássica (constante)
    a_classical = q * E / m
    v_classical[i] = v_classical[i-1] + a_classical * dt

    # Aceleração relativística
    gamma = 1 / np.sqrt(1 - (v_relativistic[i-1]/c)**2) if v_relativistic[i-1] < c else 1e10  # evitar divisão por zero
    a_rel = (q * E) / (gamma**3 * m)
    v_relativistic[i] = v_relativistic[i-1] + a_rel * dt

# Plotagem
plt.figure(figsize=(10, 6))
plt.plot(t * 1e3, v_classical / 1e8, label='Clássica', linestyle='--')
plt.plot(t * 1e3, v_relativistic / 1e8, label='Relativística')
plt.xlabel('Tempo (ms)')
plt.ylabel(r'Velocidade ($10^8$ m/s)')
plt.title('Velocidade do elétron sob um campo elétrico constante')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Próton
Foram ajustados os parâmetros, para o cálculo da velocidade do próton: 
- massa, em kg
- carga, em C

Foram também ajustados os seguintes parâmetros:
- campo elétrico: para que o mesmo ficasse positivo, considerando a carga do próton.
- tempo, pois, por ser a massa do próton maior, será mais demorado para atingir velocidades relativísticas.
- passo de tempo, devido à ordem de grandeza.
